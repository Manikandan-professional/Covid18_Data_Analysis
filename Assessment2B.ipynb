###### Department wise codekata performence (pie chart)

```python
df.Department.value_counts()
```

```python
cse=df[df['Department']=='Computer Science and Engineering']
average_performance_of_cse_department=cse['CodeKata Score'].sum()//len(cse['CodeKata Score'])

ece=df[df['Department']=='Electronics and Communication Engineering']
average_performance_of_ece_department=ece['CodeKata Score'].sum()//len(ece['CodeKata Score'])

eee=df[df['Department']=='Electronics and Electrical Engineering']
average_performance_of_eee_department=eee['CodeKata Score'].sum()//len(eee['CodeKata Score'])
data=[average_performance_of_cse_department,average_performance_of_ece_department,average_performance_of_eee_department]
```

```python
plt.figure(figsize=(16,8),dpi=80)
l=['cse','ece','eee']
plt.pie(data,explode=[0.1,0.1,0.1],labels=l)
plt.show()
```
